Develop a Python script to perform part-of-speech tagging on a short text dataset.

Import Required Libraries

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.tag import pos_tag

print("Downloading NLTK resources for English only...")
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
print("NLTK English resources ready!")
```

```python
dataset = [
    "The cat sat on the mat.",
    "She quickly ran to the store.",
    "I love programming in Python.",
    "The weather is beautiful today.",
    "Machine learning is fascinating."
]

print("Dataset loaded:")
for i, sentence in enumerate(dataset, 1):
    print(f"{i}. {sentence}")
```

```python
print("POS Tagging for dataset:")

for sentence in dataset:
    tokens = word_tokenize(sentence)
    pos_tags = pos_tag(tokens)
    
    print(f"Sentence: {sentence}")
    print("Word → POS Tag")
    for word, tag in pos_tags:
        print(f"{word} → {tag}")
    print()
```
