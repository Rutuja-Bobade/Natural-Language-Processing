```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Model
from tensorflow.keras.layers import LSTM, Embedding, Dense, TimeDistributed, Input, Bidirectional
```

```python
sentences = [
["Rutuja","Bobade","lives","in","Pune","India"],
["She","studies","IT","at","Pune","University"],
["Rutuja","attended","a","conference","organized","by","Google"],
["Sundar","Pichai","is","the","CEO","of","Google"],
["Google","has","offices","in","California","and","London"],
["Elon","Musk","founded","Tesla","and","SpaceX"],
["Tesla","is","based","in","California"],
["Engineers","from","Stanford","University","work","at","Tesla"],
["Narendra","Modi","is","the","Prime","Minister","of","India"],
["Narendra","Modi","visited","Washington"],
["Officials","from","the","United","Nations","attended","the","meeting"],
["Apple","released","a","new","iPhone"],
["Tim","Cook","is","the","CEO","of","Apple"],
["Amazon","was","founded","by","Jeff","Bezos"],
["Amazon","has","headquarters","in","Seattle"],
["Researchers","from","Harvard","University","collaborate","with","Google"],
["Microsoft","opened","a","new","office","in","Hyderabad"],
["Satya","Nadella","is","the","CEO","of","Microsoft"],
["Facebook","was","founded","by","Mark","Zuckerberg"],
["Meta","owns","Facebook","and","Instagram"]
]

labels = [
["B-PER","I-PER","O","O","B-LOC","I-LOC"],
["O","O","O","O","O","B-ORG","I-ORG"],
["B-PER","O","O","O","O","O","B-ORG"],
["B-PER","I-PER","O","O","O","O","B-ORG"],
["B-ORG","O","O","O","B-LOC","O","B-LOC"],
["B-PER","I-PER","O","B-ORG","O","B-ORG"],
["B-ORG","O","O","O","B-LOC"],
["O","O","B-ORG","I-ORG","O","O","B-ORG"],
["B-PER","I-PER","O","O","O","O","O","B-LOC"],
["B-PER","I-PER","O","B-LOC"],
["O","O","O","B-ORG","I-ORG","O","O","O"],
["B-ORG","O","O","O","O"],
["B-PER","I-PER","O","O","O","O","B-ORG"],
["B-ORG","O","O","O","B-PER","I-PER"],
["B-ORG","O","O","O","B-LOC"],
["O","O","B-ORG","I-ORG","O","O","B-ORG"],
["B-ORG","O","O","O","O","O","B-LOC"],
["B-PER","I-PER","O","O","O","O","B-ORG"],
["B-ORG","O","O","O","B-PER","I-PER"],
["B-ORG","O","B-ORG","O","B-ORG"]
]
```

```python
words = list(set(w for s in sentences for w in s))
word2idx = {w:i+1 for i,w in enumerate(words)}

tags = list(set(t for l in labels for t in l))
tag2idx = {t:i for i,t in enumerate(tags)}
idx2tag = {i:t for t,i in tag2idx.items()}
```

```python
X = [[word2idx[w] for w in s] for s in sentences]
y = [[tag2idx[t] for t in l] for l in labels]
```

```python
max_len = 6

X = pad_sequences(X, maxlen=max_len, padding="post")
y = pad_sequences(y, maxlen=max_len, padding="post")
```

```python
y = tf.keras.utils.to_categorical(y, num_classes=len(tags))
```

```python
input = Input(shape=(max_len,))

model = Embedding(input_dim=len(words)+1,
                  output_dim=64,
                  input_length=max_len)(input)

model = Bidirectional(LSTM(units=50, return_sequences=True))(model)

model = TimeDistributed(Dense(len(tags), activation="softmax"))(model)

model = Model(input, model)
```

```python
model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)
```

```python
model.fit(
    X,
    y,
    batch_size=2,
    epochs=20
)
```

```python
test_sentence = ["Rutuja", "Bobade", "works", "at", "Google"]

test_seq = [word2idx.get(w,0) for w in test_sentence]
test_seq = pad_sequences([test_seq], maxlen=max_len, padding="post")

prediction = model.predict(test_seq)
```

```python
predicted_tags = np.argmax(prediction, axis=-1)
```

```python
predicted_tags_numerical = predicted_tags[0]
```

```python
predicted_tags_string = [idx2tag[p] for p in predicted_tags_numerical]
print(predicted_tags_string)
```
