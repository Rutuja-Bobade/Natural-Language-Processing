Implement tokenization, stemming, lemmantization on a small corpus of text

# Download libraries

```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt_tab')

corpus = """Natural Language Processing (NLP) is a fascinating field of Artificial Intelligence.
It allows computers to understand, interpret, and generate human language.
Students are learning NLP techniques like t okenization, stemming, and lemmatization.
"""
```

## Tokenization
Tokenization is the process of breaking down a text into smaller units, such as words or sentences, called tokens.

```python
from nltk.tokenize import word_tokenize

tokens = word_tokenize(corpus)
```

```python
tokens
```

```python
from nltk.tokenize import sent_tokenize

sentences = sent_tokenize(corpus)
sentences
```

## Stemming



```python

from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in tokens]
print("Original Tokens:", tokens)
print("\nStemmed Words:", stemmed_words)
```

## Lemmatization

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in tokens]
print("Original Tokens:", tokens)
print(
"\nLemmatized Words:", lemmatized_words)
```

# Stop Word Removal

```python
from nltk.corpus import stopwords
nltk.download("stopwords")
```

```python
sw = set(stopwords.words("english"))

filtered_tokens = [t for t in tokens if t.lower() not in sw]

filtered_tokens
```
