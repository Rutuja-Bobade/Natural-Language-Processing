```python
import spacy

try:
    nlp = spacy.load("en_core_web_sm")
except OSError:
    print("Downloading spaCy model 'en_core_web_sm'...")
    spacy.cli.download("en_core_web_sm")
    nlp = spacy.load("en_core_web_sm")

paragraph = "The quick brown fox jumps over the lazy dog. A beautiful morning sun shines brightly. Natural language processing is a fascinating field of artificial intelligence."

doc = nlp(paragraph)

print("Noun Phrases ")
for chunk in doc.noun_chunks:
    print(chunk.text)

print(" Dependency Parsing" )
print("TOKEN\t\tDEP\t\tHEAD\t\tHEAD_POS\t\tCHILDREN")
print("-----\t\t-----\t\t-----\t\t----------\t\t---------")
for token in doc:
    children = [child.text for child in token.children]
    print(f"{token.text}\t\t{token.dep_}\t\t{token.head.text}\t\t{token.head.pos_}\t\t{children}")
```

```python
import stanza
import sys

stanza.download('en')

nlp_stanza = stanza.Pipeline(lang='en', processors='tokenize,pos,constituency')

doc_stanza = nlp_stanza(paragraph)

print("\nConstituency Parsing ")

print(f"Found {len(doc_stanza.sentences)} sentences. Displaying constituency trees:", file=sys.stderr)
for sentence_idx, sentence in enumerate(doc_stanza.sentences):
    print(f"--- Sentence {sentence_idx + 1} Constituency Tree ---")
    if sentence.constituency:
        print(str(sentence.constituency))
```
